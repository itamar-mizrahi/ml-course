<h2 dir='rtl'><u>הבנת ביטויים בוליאנים</u></h2>
<p dir='rtl'><b>מה המשמעות של הביטויים הבאים?  נסו לנחש ולאחר מכן בדקו בפייתון </b></p>


```python
print("1) True and True            ==>" , True and True)
print("2) False and True           ==>" , False and True)
print("3) True or True             ==>" , True or True)
print("4) True or False            ==>",  True or False)
print("5) not True                 ==>",  not True)
print("6) not False                ==>",  not False)
print("7) 'test' == 'test'         ==>",  "test" == "test")
print("8) True and 1 == 1          ==>",  True and 1 == 1)
print("9) False and 0 != 0         ==>",  False and 0 != 0)
print("10) 'test' == 'testing'     ==>",  "test" == "testing")
print("11) 'test' != 'testing'     ==>",  "test" != "testing")
print("12) 1 != 0 and 2 == 1       ==>",  1 != 0 and 2 == 1)
print("13) 'test' == 1             ==>",  "test" == 1)
print("14) not (True and False)    ==>",  not (True and False))
print("15) not (1 != 10 or 3 == 4) ==>",  not (1 != 10 or 3 == 4))
print("16) 8 > 0 and 8 < 10        ==>",  8 > 0 and 8 < 10)
print("17) 8%2 ==0 or 8%3 == 0     ==>",  8%2 ==0 or 8%3 == 0)
print("18) not (8 <5)              ==>",  not (8 <5))
print("19) 17 and True             ==>",  17 and True)
print("==========רגע חושבים===========")
print("=============1===============")
print("בפייתון כל מספר שהוא לא אפס נחשב כ")
print("True!!!")
print("מצד אחד זה מאוד שימושי מצד שני גמישות זו יכולה לבלבל")
print("20) 1 == 1 and not ('testing' == 1 or 1 == 0) ==>",  1 == 1 and not ("testing" == 1 or 1 == 0))
x=13; print("21) when x=13 , x<10 or x>12            ==>",  x<10 or x>12)
x=7;  print("22) when x=7 ,  x<10                    ==>",  x<10)
print("==========רגע חושבים===========")
print("=============2===============")
print("הסימון ';' משמש להפרדת פקודות באותה השורה, כך שניתן לכתוב מספר פקודות באותה השורה על ידי שימוש בסימון זה")
```

<h2 dir='rtl'><u> לולאות </u></h2>

<h3 dir='rtl'>סימולציה להטלת מטבע</h3>
<p dir='rtl'>כאשר מטילים מטבע הוגן אנו מניחים שב-1/2 ממספר ההטלות נקבל צד אחד וב-1/2 ממספר ההטלות נקבל את הצד השני.
האם הנחה זו נכונה?
על מנת לבדוק זאת, נוכל להטיל מטבע 1000 פעמים ולבדוק האם בערך ב- 500 הטלות מטבע  נקבל צד אחד וב-500  הטלות נוספות נקבל את הצד השני.
נוכל להטיל מס' רב יותר של פעמים:  10000, 100000 ויותר ולבדוק אם במחצית ממספר ההטלות קיבלנו צד אחד ובמחצית שניה קבלנו את הצד השני.
הטלת מטבע מספר רב של פעמים היא משימה לא פשוטה שדורשת זמן רב מאוד (הטלה של 1000 פעמים תיקח לנו שעות).
בעזרת כתיבת תכנית בפייתון,  נוכל לדמות הטלת מטבע כמה פעמים שנרצה ולספור את מספר הפעמים שנקבל כל צד במטבע.
על מנת לכתוב תכנית כזאת שתאפשר לנו לחזור על פקודה, במקרה שלנו הטלת מטבע מספר רב של פעמים באופן אוטומטי, נכיר מהן לולאות.</p>

<p dir='rtl'><u>דוגמא 1 - לולאת while</u></p>



```python
countdown = 10

while countdown > 0:

    print (countdown)

    countdown = countdown - 1

print ("כל הכבוד על הספירה לאחור!")
```
