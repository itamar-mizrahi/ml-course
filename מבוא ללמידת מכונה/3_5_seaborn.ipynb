קריאת נתונים מקובץ

```python
#ייבוא הנתונים למשתנה
tdf = sns.load_dataset('tips')
#הצגת המימדים - מספר שורות ומספר עמודות
print(tdf.shape)
#הצגת 10 השורות הראשונות
tdf.head(10)
```

```python
#קורלציה בין סכום החשבון לטיפ והשונות בין בנים לבנות
sns.relplot(x='total_bill', y='tip', data=tdf, hue='sex', style='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tdf)
```

<h2 dir='rtl'><u>הרחבה – בחינת קו המגמה בהתאם לשני גורמים</u></h2>
<h2 dir='rtl'>גורם 1 - סוג הארוחה - צהריים או ערב</h2>
<h2 dir='rtl'>גורם 2 - המגדר - בן או בת</h2>

```python
sns.lmplot(x='total_bill', y='tip', data=tdf,
           row='sex', col='time',
           hue='sex',
           palette='Set1',
           height=3, aspect=1.61)
```
