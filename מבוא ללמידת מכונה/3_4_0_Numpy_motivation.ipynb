<p dir = rtl>
ניצור רשימה של 1,000,000 מספרים אקראיים

```python
l = []
for i in range(10000000):
  l.append(random.random())
print(len(l))
```

```python
l[0:10]
```

<p dir = rtl>
נכתוב פעולה המקבלת רשימה של מספרים ומחזירה רשימה של ריבועי הרשימה שהתקבלה כפרמטר
</p>

```python
def square_list(l):
  for i in range(len(l)):
    l[i] = l[i]**2
  return l
```

<p dir = rtl>
נפעיל את הפעולה שכתבנו על הרשימה הארוכה שיצרנו ונמדוד את זמן הריצה
</p>

```python
tic = time.time()
l_squared = square_list(l)
toc = time.time()
print(toc - tic)
```

<p dir = rtl>
זמן הריצה יכול להשתנות מהרצה להרצה, אבל סדר הגודל הוא כ 1-1.5 שניות
<br>
נדפיס את עשרת המקומות הראשונים כדי לבדוק שהפעולה אכן בוצעה
</p>

```python
l_squared[0:10]
```

```python
def square_list_np(lnp):
  return np.square(lnp)
```

```python
lnp = np.array(l)
tic = time.time()
lnps = square_list_np(lnp)
toc = time.time()
print(toc - tic)
```
