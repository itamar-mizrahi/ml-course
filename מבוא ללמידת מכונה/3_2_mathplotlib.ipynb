<h2 dir='rtl'><u>שרטוט הפונקציה סינוס הצירים - מערכי numpy</u></h2>

```python
t = np.arange(0.0, 2.0, 0.01) #(first value, last value, step)
y1 = 3*np.sin(2 * np.pi * t)
y2 = 5*np.sin(2 * np.pi * 2* t)

plt.plot(t, y1, label='first curve')
plt.plot(t, y2, label='second curve')

plt.grid() # display grid lines
plt.title("Sine wave")
plt.xlabel("time (s)")
plt.ylabel("sine wave")

plt.legend()
plt.show()
```

<h1 dir='rtl'><b>תרגיל 1 - גרף סינוס</b></h1>
<p dir='rtl'>שרטוט גרף סינוס לפי המשוואה הבאה:</p>
<p dir='rtl'>y1 = CONST1 * sin(CONST1 * pi * arr1)</p>
<p dir='rtl'>y2 = CONST1 * sin(CONST2 * pi * arr1)</p>
<p dir='rtl'>CONST1 = מספר כלשהו</p>
<p dir='rtl'>CONST2 = מספר כלשהו</p>
<p dir='rtl'>arr1 - מערך שמתחיל מ 0 עד 2 בקפיצות של 0.01</p>

<h2 dir='rtl'><u>הצגת מספר גרפים על אותה מערכת צירים</u></h2>

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("plot 1.1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot 1.2")

plt.suptitle("all charts")
plt.show()
```

```python
rows=1
cols=2
fig, (ax1, ax2) = plt.subplots(rows, cols)

#plot 1:
ax1.set_title("plot 1.1")
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
ax1.plot(x,y)

#plot 2:
ax2.set_title("plot 1.2")
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
ax2.plot(x,y)

fig.suptitle("all charts")
```
