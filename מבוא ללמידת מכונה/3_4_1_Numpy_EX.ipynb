````python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

print (f"x = \n {x}")
print (f"\ny = \n {y}")
# Elementwise sum

print ("\nx + y =")
print(x + y) # as operator
print()

print(np.add(x, y)) # as a function
````

````python
# Elementwise difference;
print(x - y)
print()
print(np.subtract(x, y))
````

````python
# Elementwise product;
print(x * y)
print()
print(np.multiply(x, y))
````

````python
# Elementwise division;
print(x / y)
print()
print(np.divide(x, y))
````

````python
# Elementwise square root;
print(np.sqrt(x))
````

````python
# Elementwise power;
print(np.power(x, 2)) # העלאה בריבוע של כל אחד מאיברי המטריצה
````

6. מכפלה פנימית - Dot product
</b>
</h1>
<p dir=rtl>
  בכדי לעשות מכפלה פנימית של מטריצות ו\או וקטורים (באנגלית נקרא inner product או dot product) נקרא לפונקציה המובנת dot כמו בדוגמא למטה:
<p/>

````python
# מכפלה פנימית של שני וקטורים
# שני הוקטורים מוכרחים להיות בעלי אותו מספר מימדים

# נגדיר שני וקטורים
v = np.array([10,5]) # וקטור עם 2 מימדים
w = np.array([2,1]) # וקטור עם 2 מימדים


# את המכפלה אפשר לעשות על ידי כל אחד מהצורות הבאות
print(v)
print(w)
print(v.dot(w))
print(np.dot(v, w))

````

<p dir = rtl>
מאחר ופעולת המכפלה בין וקטורים ומטריצות נמצאת בשימוש נרחב מאוד, החל מגרסת פיתון 3.5 (נכנסה לעבודה ב2016) במקום לקרוא לפונקציה המובנת  dot ניתן להשתמש באופרטור ״@״ על מנת לבצע מכפלה זו.
</p>


````python
print(v)
print(w)
print(v @ w)
````

<p dir=rtl>
כיום משתמשים ב״@״ באופן כמעט בלעדי. עם זאת, בדוגמאות רבות הקיימות ברשת עוד לפני 2016 תמצאו כתיב של dot  
</p>


````python
# מכפלה פנימית של וקטור ומטריצה
#מימד הוקטור חייב להיות זהה למימד העמודות של המטריצה

A = np.array([[1,2],[3,4], [5,5],[6,0]]) # מטריצה  4 על 2
print(f"A = \n{A}")
print(f"v = {v}")
print(f"w = {w}")


print(f"A @ w = {A@w}")
print(f"A @ v = {A@v}")
````

````python
A = np.array([[1,2],[3,4], [5,5],[6,0]]) # מטריצה  4 על 2
print("A =")
print (A)
print ("\nA.shape = ")
print(A.shape)
print("\nA.T =")
print (A.T)
print ("\nA.T.shape = ")
print(A.T.shape)
````

````python
v = np.array([[1,2,3]])
print(v )
print("transpose\n", v.T)
````

 בnumpy  יש אבחנה בין וקטור למטריצה חד ממדית.
<br>
הshape  של וקטור בעל 5 אלמנטים יראה כך: (5,)
<br>
בעוד במטריצה חד ממדית הshape  יהיה:  .(5,1)

````python
c = np.array([3,5,7])
print(c.shape)
print(c)
d = np.array([[3],[5],[7]])
print(d.shape)
print(d)
````

<p dir=rtl>
 אבחנה זו חשובה מאחר וכאשר מנסים להכפיל וקטור במטריצה, כל עוד אחד מממדי המטריצה תואם למימד הוקטור, numpy  יבצע את הכפלה.
<br>
לעומת זאת, אם ננסה להכפיל מטריצה חד ממדית, נצטרך לדאוג לעשות לה transpose בהתאם לצורת המטריצה הדו ממדית בה אנו רוצים לכפול.
